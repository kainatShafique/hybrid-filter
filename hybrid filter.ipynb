```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.ndimage import gaussian_filter, sobel
from skimage import data, color

# Load a sample image
image = color.rgb2gray(data.astronaut())

# Apply a Gaussian (low-pass) filter
low_pass = gaussian_filter(image, sigma=5)

# Apply a Sobel (high-pass) filter to extract edges
sobel_x = sobel(image, axis=0)
sobel_y = sobel(image, axis=1)
high_pass = np.hypot(sobel_x, sobel_y)

# Combine the low-pass and high-pass filters
hybrid_image = low_pass + high_pass

# Visualize the results
fig, axes = plt.subplots(1, 4, figsize=(15, 5))
axes[0].imshow(image, cmap='gray')
axes[0].set_title("Original Image")
axes[0].axis('off')

axes[1].imshow(low_pass, cmap='gray')
axes[1].set_title("Low-Pass Filtered")
axes[1].axis('off')

axes[2].imshow(high_pass, cmap='gray')
axes[2].set_title("High-Pass Filtered")
axes[2].axis('off')

axes[3].imshow(hybrid_image, cmap='gray')
axes[3].set_title("Hybrid Image")
axes[3].axis('off')

plt.tight_layout()
plt.show()
```
